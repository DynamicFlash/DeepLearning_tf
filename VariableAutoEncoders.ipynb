```python
import tensorflow as tf
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
```

Get Dataset

```python
from tensorflow.keras.datasets import fashion_mnist

(X_train, y_train), (X_test, y_test) = fashion_mnist.load_data()
```

Identifying the classes and volume for each class

```python
all_classes, vol_classes = np.unique(y_train, return_counts= True)
```

Helper function to get indexes of each class such that we can train our Autoencoder specifically to the class

```python
def get_indexes(array, classes):
    #create an empty diction
    indexes = {}

    #Populate indexes for each class
    # Complexicity is O(n*m)
    for feature in classes:
        temp_list = [x for x in range(len(array)) if array[x] == feature]
        indexes[feature] = temp_list
        temp_list = None

    return indexes

indexes = get_indexes(y_train, all_classes)
```

Segregation of Indexes

```python
class_0 = X_train[indexes[0]]
class_1 = X_train[indexes[1]]
class_2 = X_train[indexes[2]]
class_3 = X_train[indexes[3]]
class_4 = X_train[indexes[4]]
class_5 = X_train[indexes[5]]
class_6 = X_train[indexes[6]]
class_7 = X_train[indexes[7]]
class_8 = X_train[indexes[8]]
class_9 = X_train[indexes[9]]
```

Sampling each class

```python
import random

fig = plt.figure(figsize=(10, 7))

rows = 5
columns = 2

for image_no in range(0,len(all_classes)):

    fig.add_subplot(rows, columns, image_no + 1)
  
    # showing image
    plt.imshow(np.reshape(X_train[random.choices(indexes[image_no])], (28,28)))
    plt.axis('off')
    plt.title(str(image_no+1))
    plt.tight_layout()
```
